```python
import pandas as pd
import numpy as np
import duckdb
import warnings
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from scipy.stats import randint, uniform
import matplotlib.pyplot as plt
import xgboost as xgb
import joblib
import warnings

pd.set_option('display.max_columns', None)
warnings.filterwarnings("ignore")

categories = ['PTS', 'AST', 'REB', 'PR', 'PA', 'RA', 'PRA', 'TPM', 'STL', 'BLK', 'STL_BLK']
con = duckdb.connect(database=":memory:")
year = 2025
```

# Feature Engineering

```python
df_actuals = pd.read_csv(f"../tables/{year}/parlay_actuals.csv")
# df_stats = pd.read_csv(f"../tables/{year}/parlay_stats.csv")
# stats_cols = df_stats.columns.tolist()
# for stat in categories:
#     stats_cols.append(stat)
# df_actuals = df_actuals[stats_cols]
df = df_actuals.copy()
tgt_stat = "PTS"
df['MP_L1'] = df.groupby(['Team', 'Player'])['MP'].shift(1)

df['MP_L3'] = (
    df.groupby(['Team', 'Player'])['MP']
      .rolling(3)
      .mean()
      .shift(1)
      .reset_index(level=[0,1], drop=True)
)

df['MP_L5'] = (
    df.groupby(['Team', 'Player'])['MP']
      .rolling(5)
      .mean()
      .shift(1)
      .reset_index(level=[0,1], drop=True)
)
df = df[['Date', 'Team', 'B2B', 'Player', 'Pos', 'MP_L1', 'MP_L3', 'MP_L5', 'Opp', 
         f'Off_{tgt_stat}', f'Off_L5_{tgt_stat}', f'AVG_{tgt_stat}_H2H', 
         f'Def_{tgt_stat}', f'Def_L5_{tgt_stat}', f'{tgt_stat}_line',
         f'{tgt_stat}_Hit_Pct', f'{tgt_stat}_P_Diff', 'Spread', 'Total', tgt_stat]]
df
```

# Baseline Model

```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].astype('category')
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
n = len(df)
train_end = int(0.7 * n)
val_end   = int(0.85 * n)

train_df = df.iloc[:train_end]
val_df   = df.iloc[train_end:val_end]
test_df  = df.iloc[val_end:]

print('Train:', len(train_df), '/ Validation:', len(val_df), '/ Test:', len(test_df))

feature_cols = [col for col in df.columns 
                if col not in ['Date', 'Team', 'Player', tgt_stat]]

X_train, y_train = train_df[feature_cols], train_df[tgt_stat]
X_val,   y_val   = val_df[feature_cols],   val_df[tgt_stat]
X_test,  y_test  = test_df[feature_cols],  test_df[tgt_stat]


model = XGBRegressor(
    enable_categorical=True,
    n_estimators=300,    # number of trees
    learning_rate=0.05,  # step size
    max_depth=6,         # complexity
    subsample=0.8,       # row sampling
    colsample_bytree=0.8,
    objective='reg:squarederror',
    random_state=42
)

model.fit(
    X_train, y_train,
    eval_set=[(X_val, y_val)],
    verbose=False
)

preds = model.predict(X_test)

rmse = np.sqrt(mean_squared_error(y_test, preds))
mae = mean_absolute_error(y_test, preds)
r2 = r2_score(y_test, preds)

print("RMSE:", rmse)
print("MAE:", mae)
print("R²:", r2)
```
