```python
def partition_save_df(df, file_path):
    os.makedirs(os.path.dirname(file_path), exist_ok=True)
    try:
        df_old = pd.read_csv(file_path)
        df_old['Date'] = pd.to_datetime(df_old['Date']).dt.strftime('%Y-%m-%d')
        df_old = df_old[df_old['Date'] != now]
    except FileNotFoundError:
        df_old = pd.DataFrame()
    df_combined = pd.concat([df_old, df], ignore_index=True)
    df_combined['Date'] = pd.to_datetime(df_combined['Date'], format='mixed', errors='coerce')
    df_combined['Date'] = df_combined['Date'].dt.strftime("%m/%d/%Y")
    
    if file_path.split("/")[-1] in ['parlay_stats.csv', 'parlay_lines.csv', 'parlay_actuals.csv']:
        df_combined = df_combined[~df_combined[['Date', 'Team', 'Player']].duplicated(keep='last')]
    
    df_combined.to_csv(file_path, index=False)
    print(file_path, 'saved!')
```

```python
def travel_distance(team_prev, team_curr):
    loc_prev = arenas[team_prev]
    loc_curr = arenas[team_curr]
    return haversine(loc_prev, loc_curr)  # returns kilometers
```
