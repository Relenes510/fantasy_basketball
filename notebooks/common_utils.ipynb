```python
def partition_save_df(df, file_path):
    try:
        df_old = pd.read_csv(file_path)
        df_old['Date'] = pd.to_datetime(df_old['Date']).dt.strftime('%Y-%m-%d')
        df_old = df_old[df_old['Date'] != now]
    except FileNotFoundError:
        df_old = pd.DataFrame()
    df_combined = pd.concat([df_old, df], ignore_index=True)
    df_combined['Date'] = pd.to_datetime(df_combined['Date'], format='mixed', errors='coerce')
    df_combined['Date'] = df_combined['Date'].dt.strftime("%m/%d/%Y")
    df_combined.to_csv(file_path, index=False)
    print(file_path, 'saved!')
```
